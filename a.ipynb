```python
import pandas as pd

# Load your dataset from a CSV file
df = pd.read_csv("google/filtgoogleplay.csv")

# Calculate the median of the 'Rating' column
median_Rating = df['Rating'].median()

# Replace Ratings greater than 5 with the median
df['Rating'] = df['Rating'].apply(lambda x: x if x <= 5 else median_Rating)

# Save the modified data back to a CSV file if needed
df.to_csv("google/filtgoogleplay.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Convert 'Reviews' and 'Rating' columns to numeric
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Clean 'Installs' column and convert to numeric
df['Installs'] = df['Installs'].apply(lambda x: re.sub(r'[^\d]+', '', str(x)))
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')

# Sort df by 'Installs' in descending order and select the top 30 apps
top_30_apps = df.nlargest(30, 'Installs')

# Plotting the df
plt.figure(figsize=(12, 8))
plt.bar(top_30_apps['App'], top_30_apps['Installs'],
        color='skyblue', label='Installs')
plt.bar(top_30_apps['App'], top_30_apps['Reviews'],
        color='lightgreen', label='Reviews')
plt.plot(top_30_apps['App'], top_30_apps['Rating'],
         color='orange', marker='o', label='Rating')

plt.title('Top 30 Apps: Installs, Reviews, and Ratings')
plt.xlabel('Apps')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
